```python
all_grades = school_all_data["grade"].unique()
reorderList = [0,3,2,1]
all_grades = [all_grades[i] for i in reorderList] 
#all_grades
```

```python
def getMathData():
    results = {}
    math_results_Avg_School = {}
    
    for school in school_names:
        math_results_Avg_School = {}
        results.update({school:math_results_Avg_School})
        
        for grade in all_grades:
            
            math_results = school_all_data.loc[(school_all_data["school_name"] == (school)) & 
                                   (school_all_data["grade"] == (grade)), :]

            math_results_Avg_School.update({grade:(math_results["math_score"].mean())})

    return results
```

```python
mathResultsPerGrade = getMathData()
#print(mathResultsPerGrade)
```

```python
results_math_df = pd.DataFrame.from_dict(mathResultsPerGrade, orient='index').rename_axis("School").sort_values("School")
results_math_df
```

```python
def getReadingData():
    results = {}
    reading_results_Avg_School = {}
    
    for school in school_names:
        reading_results_Avg_School = {}
        results.update({school:reading_results_Avg_School})
        
        for grade in all_grades:
            
            reading_results = school_all_data.loc[(school_all_data["school_name"] == (school)) & 
                                   (school_all_data["grade"] == (grade)), :]
            
            reading_results_Avg_School.update({grade:(reading_results["reading_score"].mean())})
    
    return results
```

```python
readingResultsPerGrade = getReadingData()
```

```python
results_reading_df = pd.DataFrame.from_dict(readingResultsPerGrade, orient='index').rename_axis("School").sort_values("School")
results_reading_df
```
